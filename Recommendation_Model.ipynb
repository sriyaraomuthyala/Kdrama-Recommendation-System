```python
drama['Keywords'] = drama['Keywords'].str.replace(',,', ',', regex=False)
```

```python
drama['Genre'] = drama['Genre'].apply(lambda x: [item.strip() for item in x.split(',')])
drama['Keywords'] = drama['Keywords'].apply(lambda x: [item.strip() for item in x.split(',')])
drama['Main_Cast'] = drama['Main_Cast'].apply(lambda x: [item.strip() for item in x.split(',')])
```

```python
def convert(obj):
    l = []
    counter = 0
    for i in obj:
        if counter != 3:
            l.append(i)
            counter+=1
        else:
            break
    return l
```

```python
drama.Main_Cast = drama['Main_Cast'].apply(convert)
```

```python
drama['Keywords'] = drama['Keywords'].apply(lambda x:[i.replace(' ', '') for i in x])
drama['Main_Cast'] = drama['Main_Cast'].apply(lambda x:[i.replace(' ', '') for i in x])
drama['Genre'] = drama['Keywords'].apply(lambda x:[i.replace(' ', '') for i in x])
```

```python
drama['tags'] = drama['Sinopsis']+drama['Genre']+drama['Keywords']+drama['Main_Cast']
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    s = []
    for i in text.split():
        s.append(ps.stem(i))
    return " ".join(s)
```

```python
stem('beautiful beautifully beauty')
```

```python
new_drama_data['tags'] = new_drama_data['tags'].apply(stem)
```

```python
new_drama_data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(stop_words='english',max_features=5000,analyzer='word')
vectorizer
```

```python
vectors = vectorizer.fit_transform(new_drama_data['tags']).toarray()
vectors
```

```python
vectorizer.get_feature_names_out()
```

```python
len(vectorizer.get_feature_names_out())
```

```python
corp =vectorizer.get_feature_names_out()
for i in range(5000):
    print (corp[i])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
sorted(list(enumerate(similarity[0])),reverse = True , key=lambda x:x[1])[1:6]
```

```python
def recommend(drama_name):
    drama_index = new_drama_data[new_drama_data['Name']== drama_name].index[0]
    distance = similarity[drama_index]
    drama_list = sorted(list(enumerate(distance)),reverse = True , key=lambda x:x[1])[1:6]

    for i in drama_list:
        print(new_drama_data.iloc[i[0]].Name)
        
```

```python
recommend('Weak Hero Class 1')
```

```python
new_drama_data.iloc[658].Name
```

```python
recommend('Hospital Playlist')
```
