```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
```

```python
import numpy
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.91mobiles.com/entertainment/best-family-web-series"
```

```python
r = requests.get(url=url)
```

```python
r
```

```python
soup = BeautifulSoup(r.content,'html.parser')
soup
```

```python
result = soup.find_all('div', class_='result_items')
```

```python
result
```

```python
series_containers = soup.find_all('div', class_='pro_item card_box bg-inner-c txt-white')

#Creating an empty list
series_data = []

# Extract the required details from each series container
for container in series_containers:
    # Extract the data-label value
    data_label = container.find('div', class_='target_link ga_tracking').get('data-label', 'N/A')
    series_name = data_label.split(':/entertainment/best-family-web-series')[0]
    
    # Extract the Hindi text and date
    language = container.find('p', class_='d-in-block f-s-m').get_text(strip=True)
    language = language.split(" |")[0]

    #Extracting the genre
    genre_label = container.find('label', string='Genres')
    genre = genre_label.find_next_sibling('p', class_='two-l-trun').get_text(strip=True) if genre_label else 'N/A'
    
    # Extract the cast
    cast = container.find('p', class_='two-l-trun').get_text(strip=True)
    
    # Extract the image URL
    img_tag = container.find('img', class_='lazy d-block')
    img_url = img_tag['data-src'] if img_tag else 'N/A'
    
    series_data.append({
            "series_name": series_name,
            "language": language,
            "genre": genre,
            "cast": cast,
            "img": img_url
        })

    # Print the extracted details
    print(f"Series Name: {series_name}")
    print(f"Language: {language}")
    print(f"Cast: {cast}")
    print(f"Genre: {genre}")
    print(f"Image URL: {img_url}")
    print('-' * 40)
    
else:
    print(f"Failed to retrieve the webpage. Status code: {r.status_code}")
```
